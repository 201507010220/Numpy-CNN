```python
from CNN.utils import initializeFilterNormal, initializeWeightsXavier, nanargmax
from CNN.forward import maxpool, softmax, logloss


import numpy as np
```

```python
filt1 = initializeFilterNormal(8,5,1)
filt2 = initializeFilterNormal(8,5,8)
theta3 = initializeWeightsXavier(10, 800)
bias1 = np.zeros((8,1), dtype = np.float16)
bias2 = np.zeros((8,1), dtype = np.float16)
bias3 = np.zeros((10, 1), dtype = np.float16)

sample_image = np.random.normal(loc = 0, scale = 1, size = (1, 28, 28))
sample_label = np.zeros((10, 1))
sample_label[5] = 1
```

```python
def ConvNet(image, label, filt1, filt2, bias1, bias2, theta3, bias3):
    
    
    (l, w, w) = image.shape # input dims
    l1, l2 = len(filt1), len(filt2) ## number of filters for each conv layer
    (_, f, _) = filt1[0].shape #filter size
    
    w1 = w - f + 1 # output shape of first conv layer
    w2 = w1 - f + 1 # output shape of second conv layer
    
    ####################################################
    ################## Forward Step ####################
    ####################################################
    
    # Convolution Operation #1
    conv1 = np.zeros((l1, w1, w1)) 
    
    for jj in range(l1):
        for x in range(w1):
            for y in range(w1):
                conv1[jj, x, y] = np.sum(filt1[jj] * image[:, x:x+f, y:y+f]) + bias1[jj]
                
    ## ReLU
    conv1[conv1<=0] = 0
    
    # Convolution Opetation #2
    conv2 = np.zeros((l2, w2, w2))
    
    for jj in range(l2):
        for x in range(w2):
            for y in range(w2):
                conv2[jj, x, y] = np.sum(filt2[jj] * conv1[:, x:x+f,y:y+f]) + bias2[jj]
                
    
    ## ReLU
    conv2[conv2 <= 0] = 0
    
    # Maxpooling operation
    pooled_layer = maxpool(conv2, 2, 2)
    
    # Fully connected operation
    fc = pooled_layer.reshape((l2 * w2//2 * w2//2, 1))
    
    out = np.dot(theta3, fc) + bias3
    
    # Softmax
    probs = softmax(out)
    
    ####################################################
    ###################### Loss ########################
    ####################################################    
    
    loss = logloss(label, probs)
    
    if np.argmax(label) == np.argmax(probs):
        acc = 1
    else:
        acc = 0
        
    ####################################################
    ################## Backward Step ###################
    ####################################################
    
    dout = probs - label
    
    dtheta3 = dout.dot(fc.T)
    
    dbias3 = np.sum(dout, axis = 1).reshape(10,1)
    
    dfc = theta3.T.dot(dout)
    
    dpool = dfc.reshape((l2, w2//2, w2//2))
    
    dconv2 = np.zeros((l2, w2, w2))
    
    for jj in range(l2):
        curr_x = out_x = 0
        while curr_x + f <= w2:
            curr_y = out_y = 0
            while curr_y + f <= w2:
                (a, b) = nanargmax(conv2[jj, curr_x:curr_x+f, curr_y:curr_y+f])
                dconv2[jj, curr_x+a, curr_y+b] = dpool[jj, out_x, out_y]
                
                curr_y += 2
                out_y += 1
            curr_x += 2
            out_x += 1
            
    dconv2[conv2 <= 0] = 0
    
    dfilt2 = np.zeros((l2, l1, f, f))
    dfilt1 = np.zeros((l1, l, f, f))
    dbias2 = np.zeros((l2,1))
    dbias1 = np.zeros((l1,1))
        
    dconv1 = np.zeros((l1, w1, w1))
    
    for jj in range(l2):
        for x in range(w2):
            for y in range(w2):
                dfilt2[jj] += dconv2[jj, x, y] * conv1[:, x:x+f, y:y+f]
                dconv1[:, x:x+f, y:y+f] += dconv2[jj, x, y] * filt2[jj] 
        dbias2[jj] = np.sum(dconv2[jj])
        
    dconv1[conv1<=0] = 0
    
    for jj in range(l1):
        for x in range(w1):
            for y in range(w1):
                dfilt1[jj] += dconv1[jj, x, y] * image[:, x:x+f, y:y+f]
        dbias1[jj] = np.sum(dconv1[jj])
    
    return dfilt1, dfilt2, dtheta3, dbias1, dbias2, dbias3, loss, acc
```

```python
dfilt1, dfilt2, dtheta3, dbias1, dbias2, dbias3, loss, acc = ConvNet(sample_image, sample_label, filt1, filt2, bias1, bias2, theta3, bias3)
```

```python
def momentumSGD(batch, LEARNING_RATE, w, l, MU, filt1, filt2, bias1, bias2, theta3, bias3, cost, acc ):

    
    
    X = batch[:,0:-1]
    X = X.reshape(len(batch), l, w, w)
    Y = batch[:,-1]

    l1, l2 = len(filt1), len(filt2)
    (_, f, _ ) = filt2[0].shape
    num_classes = 10
    
    n_correct=0
    cost_ = 0
    batch_size = len(batch)
    
    # initialize GD params
    dfilt1 = np.zeros((l1, l, f, f))
    dfilt2 = np.zeros((l2, l1, f, f))
    dtheta3 = np.zeros(theta3.shape)
    dbias1 = np.zeros((l1, 1))
    dbias2 = np.zeros((l2, 1))
    dbias3 = np.zeros(bias3.shape)
    # initialize momentum params
    v1 = np.zeros((l1, l, f, f))
    v2 = np.zeros((l2, l1, f, f))
    v3 = np.zeros(theta3.shape)
    bv1 = np.zeros((l1, 1))
    bv2 = np.zeros((l2, 1))
    bv3 = np.zeros(bias3.shape)
    
    for i in range(batch_size):
        x = X[i]
        y = np.eye(num_classes)[int(Y[i])].reshape(num_classes, 1)
        
        # Collect Gradients
        
        [dfilt1_, dfilt2_, dtheta3_, dbias1_, dbias2_, dbias3_, curr_cost, acc_] = ConvNet(x, y, filt1, filt2, bias1, bias2, theta3, bias3,)
        
        dfilt1+=dfilt1_
        dbias1+=dbias1_
        dfilt2+=dfilt2_
        dbias2+=dbias2_
        dtheta3+=dtheta3_
        dbias3+=dbias3_

        cost_+=curr_cost
        n_correct+=acc_

    v1 = MU*v1 - LEARNING_RATE*dfilt1/batch_size
    filt1 += v1
    # filt1[j] -= LEARNING_RATE*dfilt1[j]/batch_size
    bv1 = MU*bv1 -LEARNING_RATE*dbias1/batch_size
    bias1 += bv1
   
    v2 = MU*v2 -LEARNING_RATE*dfilt2/batch_size
    filt2 += v2
    # filt2[j] += -LEARNING_RATE*dfilt2[j]/batch_size
    bv2 = MU*bv2 -LEARNING_RATE*dbias2/batch_size
    bias2 += bv2
    
    v3 = MU*v3 - LEARNING_RATE*dtheta3/batch_size
    theta3 += v3
    # theta3 += -LEARNING_RATE*dtheta3/batch_size
    bv3 = MU*bv3 -LEARNING_RATE*dbias3/batch_size
    bias3 += bv3

    cost_ = cost_/batch_size
    cost.append(cost_)
    accuracy = float(n_correct)/batch_size
    acc.append(accuracy)

    return [filt1, filt2, bias1, bias2, theta3, bias3, cost, acc]
```

```python
import gzip

def extract_data(filename, num_images, IMAGE_WIDTH):
    """Extract the images into a 4D tensor [image index, y, x, channels].
    Values are rescaled from [0, 255] down to [-0.5, 0.5].
    """
    print('Extracting', filename)
    with gzip.open(filename) as bytestream:
        bytestream.read(16)
        buf = bytestream.read(IMAGE_WIDTH * IMAGE_WIDTH * num_images)
        data = np.frombuffer(buf, dtype=np.uint8).astype(np.float32)
        data = data.reshape(num_images, IMAGE_WIDTH*IMAGE_WIDTH)
        return data

def extract_labels(filename, num_images):
    """Extract the labels into a vector of int64 label IDs."""
    print('Extracting', filename)
    with gzip.open(filename) as bytestream:
        bytestream.read(8)
        buf = bytestream.read(1 * num_images)
        labels = np.frombuffer(buf, dtype=np.uint8).astype(np.int64)
    return labels
```

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
from matplotlib import gridspec
import pickle
import time
import random


#####################################################################################################################################
################################################## ------ START HERE --------  ######################################################
##################################### ---------- CONVOLUTIONAL NEURAL NETWORK ---------------  ######################################
################ ----ARCHITECTURE PROPOSED : [INPUT - CONV1 - RELU - CONV2 - RELU- MAXPOOL - FC1 - OUT]---- #########################
#####################################################################################################################################


## Hyperparameters
NUM_OUTPUT = 10
LEARNING_RATE = 0.01	#learning rate
IMG_WIDTH = 28
IMG_DEPTH = 1
FILTER_SIZE=5
NUM_FILT1 = 8
NUM_FILT2 = 8
BATCH_SIZE = 20
NUM_EPOCHS = 2	 # number of iterations
MU = 0.95

PICKLE_FILE = 'output.pickle'
# PICKLE_FILE = 'trained.pickle'


## Data extracting
m =10000
X = extract_data('t10k-images-idx3-ubyte.gz', m, IMG_WIDTH)
y_dash = extract_labels('t10k-labels-idx1-ubyte.gz', m).reshape(m,1)
X-= int(np.mean(X))
X/= int(np.std(X))
test_data = np.hstack((X,y_dash))


m =50000
X = extract_data('train-images-idx3-ubyte.gz', m, IMG_WIDTH)
y_dash = extract_labels('train-labels-idx1-ubyte.gz', m).reshape(m,1)
print(np.mean(X), np.std(X))
X-= int(np.mean(X))
X/= int(np.std(X))
train_data = np.hstack((X,y_dash))

np.random.shuffle(train_data)

NUM_IMAGES = train_data.shape[0]

## Initializing all the parameters

filt1 = initializeFilterNormal(NUM_FILT1, FILTER_SIZE, IMG_DEPTH, SCALE=1.0)
bias1 = np.zeros((NUM_FILT1, 1))

filt2 = initializeFilterNormal(NUM_FILT2, FILTER_SIZE, NUM_FILT1, SCALE=1.0)
bias2 = np.zeros((NUM_FILT2, 1))
   
w1 = IMG_WIDTH-FILTER_SIZE+1
w2 = w1-FILTER_SIZE+1
theta3 = initializeWeightsXavier(NUM_OUTPUT, (w2//2)*(w2//2)*NUM_FILT2)
bias3 = np.zeros((NUM_OUTPUT,1))

cost = []
acc = []

print("Learning Rate:"+str(LEARNING_RATE)+", Batch Size:"+str(BATCH_SIZE))

## Training start here

for epoch in range(0,NUM_EPOCHS):
    np.random.shuffle(train_data)
    batches = [train_data[k:k + BATCH_SIZE] for k in range(0, NUM_IMAGES, BATCH_SIZE)]
    
    for x,batch in enumerate(batches):
        out = momentumSGD(batch, LEARNING_RATE, IMG_WIDTH, IMG_DEPTH, MU, filt1, filt2, bias1, bias2, theta3, bias3, cost, acc)
        [filt1, filt2, bias1, bias2, theta3, bias3, cost, acc] = out
        epoch_acc = round(np.sum(acc[int(epoch*NUM_IMAGES/BATCH_SIZE):])/(x+1),2)

        per = float(x+1)/len(batches)*100
        print("Epoch:"+str(round(per,2))+"% Of "+str(epoch+1)+"/"+str(NUM_EPOCHS)+", Cost:"+str(cost[-1])+", B.Acc:"+str(acc[-1]*100)+", E.Acc:"+str(epoch_acc))

        


## saving the trained model parameters
with open(PICKLE_FILE, 'wb') as file:
    pickle.dump(out, file)
```

```python
def predict(image, filt1, filt2, bias1, bias2, theta3, bias3):

    (l,w,w)=image.shape
    (l1,f,f) = filt2[0].shape
    l2 = len(filt2)
    w1 = w-f+1
    w2 = w1-f+1
    conv1 = np.zeros((l1,w1,w1))
    conv2 = np.zeros((l2,w2,w2))
    for jj in range(0,l1):
        for x in range(0,w1):
            for y in range(0,w1):
                conv1[jj,x,y] = np.sum(image[:,x:x+f,y:y+f]*filt1[jj])+bias1[jj]
    conv1[conv1<=0] = 0 #relu activation
    ## Calculating second Convolution layer
    for jj in range(0,l2):
        for x in range(0,w2):
            for y in range(0,w2):
                conv2[jj,x,y] = np.sum(conv1[:,x:x+f,y:y+f]*filt2[jj])+bias2[jj]
    conv2[conv2<=0] = 0 # relu activation
    ## Pooled layer with 2*2 size and stride 2,2
    pooled_layer = maxpool(conv2, 2, 2)	
    fc1 = pooled_layer.reshape(((w2//2)*(w2//2)*l2,1))
    out = theta3.dot(fc1) + bias3	#10*1
    eout = np.exp(out, dtype=np.float)
    probs = eout/sum(eout)
    
    return np.argmax(probs), np.max(probs)
```

```python
## Opening the saved model parameter
pickle_in = open(PICKLE_FILE, 'rb')
out = pickle.load(pickle_in)

[filt1, filt2, bias1, bias2, theta3, bias3, cost, acc] = out


## Plotting the cost and accuracy over different background
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 1]) 
ax0 = plt.subplot(gs[0])
line0, = ax0.plot(cost, color='r')
ax1 = plt.subplot(gs[1], sharex = ax0)
line1, = ax1.plot(acc, color='g', linestyle='--')
plt.setp(ax0.get_xticklabels(), visible=False)
ax0.legend((line0, line1), ('Loss', 'Accuracy'), loc='upper right')
# remove vertical gap between subplots
plt.subplots_adjust(hspace=.0)
plt.show(block=False)

## Computing Test accuracy
X = test_data[:,0:-1]
X = X.reshape(len(test_data), IMG_DEPTH, IMG_WIDTH, IMG_WIDTH)
y = test_data[:,-1]
corr = 0
print("Computing accuracy over test set:")

t = tqdm_notebook(range(len(test_data)), leave=True)

for i in t:
    x = X[i]
    pred, prob = predict(x, filt1, filt2, bias1, bias2, theta3, bias3)
    if pred==y[i]:
        corr+=1
    
    t.set_description("Acc:"+str(float(corr/(i+1))*100)+"%")

        
```
